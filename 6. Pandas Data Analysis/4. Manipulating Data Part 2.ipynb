#### Remove the $ , sign from Price column as we can plot a graph
  - We use str property and replace function to change price value into a integer type
  - ('[\$\,\.]', '')   : '[\$\,\.]' This part show that Price value may have dollar, comma sign and it will be replaced with '' empty space 
  - astype() : This function change values into a integer. Howevever, obtained entries are still a string object.

#### Note
 - Always Check the updated column before assign into a original dataframe column
 - As this line car_sales["Price"].str.replace('[\$\,\.]', '').astype(int) retrieve the record that we try to achieve after removing the sign from dataframe car_sales inside the Price column.
 - This line car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '').astype(int) show that record updated and assigned into orginal dataframe car_cales inside the Price column.

```python
#dataframe["column name"] = dataframe["column name"].str.replace('[\$\,\.]', '').astype(int)
car_sales["Price"].str.replace('[\$\,\.]', '').astype(int)
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '').astype(int)
car_sales["Price"]
```

```python
'''
car_sales_missing_dataset["Odometer"].fillna(car_sales_missing_dataset["Odometer"].mean())

.fillna() function : uses to fill the Nan values into a dataframe column.

car_sales_missing_dataset["Odometer"].mean() :  mean() function calculate the mean function of dataframe car_sales_missing_dataset
having column "Odometer" that will be 92302.66666666667. This values will be assigned to Nan entries inside the dataframe car_sales_missing_dataset having column "Odometer".

'''
car_sales_missing_dataset["Odometer"].fillna(car_sales_missing_dataset["Odometer"].mean())
```

### lambda function
  - lambda function provide easiness to apply any condition into a dataframe column.

```python
car_sales["Odometer (KM)"]
```

```python
"""
car_sales["Odometer (KM)"] : fetch all the records / rows from the dataframe car_sales having column Odometer (KM).
then apply the lambda function.

.apply(lambda x: x / 1.6)  : This fuction works like a for loop here
 x:x /1.6 means : fetch all the record / column data from {Odometer (KM)} column and divide it with 1.6 number

"""

car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)
```
