# Standard Deviation and Variance

```python
h_var_array = np.array([1, 100, 200, 300, 4000, 5000])
l_var_array = np.array([2, 4, 6, 8, 10])
```

```python
np.var(h_var_array) , np.var(l_var_array)
```

```python
np.std(h_var_array), np.std(l_var_array)
```

```python
np.mean(h_var_array), np.mean(l_var_array)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(h_var_array)
plt.show()
```
