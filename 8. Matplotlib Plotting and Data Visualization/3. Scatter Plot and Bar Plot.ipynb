### Numpy Figure

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some Data
x = np.linspace(0, 10, 100)   # Return evenly spaced numbers over a specified interval.
x
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**4)
```

```python
# Scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Other plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot using a dictionary
coldrink_prices = {"Pepsi": 60,
                   "Spride": 55,
                   "Dew": 50 
}
coldrink_prices
fig, ax = plt.subplots()
ax.bar(coldrink_prices.keys(), coldrink_prices.values()) # X, Y
```

```python
# Make a plot using a dictionary
coldrink_prices = {"Pepsi": 60,
                   "Spride": 55,
                   "Dew": 50 
}
coldrink_prices
fig, ax = plt.subplots()
ax.bar(coldrink_prices.keys(), coldrink_prices.values()) # X, Y
ax.set(title=" Cold Drink Prices",
      ylabel = "Price",
      xlabel ="Items")
```
