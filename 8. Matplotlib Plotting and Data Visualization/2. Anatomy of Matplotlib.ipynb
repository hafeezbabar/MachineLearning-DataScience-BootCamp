```python
%matplotlib inline
import matplotlib.pyplot as plt
 
# 1. Generate the data    
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]


# 2. set plot
fig, ax =plt.subplots(figsize = (10,10))


# 3. Plot the data
ax.plot(x, y)


# 4. Customize plot
ax.set(title = "Simple plt",
      xlabel = "x-axis",
      ylabel = "y-label")

# 5. Save and show (save the whole figure)
fig.savefig("C:/Users/hafee/OneDrive/Desktop/MachineLearningDataScienceBootCamp/sample_figure.png")
```
