```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make a plot using a dictionary
coldrink_prices = {"Pepsi": 60,
                   "Spride": 55,
                   "Dew": 50 
}
coldrink_prices
fig, ax = plt.subplots()
ax.barh(list(coldrink_prices.keys()), list(coldrink_prices.values())) # X, Y
```

```python
# Generate some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

## Two ways for sub plotting

```python
# Subplot 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                            ncols=2,
                                            figsize = (10,5))
```

```python
# Subplot 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                            ncols=2,
                                            figsize = (10,5))
ax1.plot(x, x /2)
```

```python
# Subplot 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                            ncols=2,
                                            figsize = (10,5))

ax1.plot(x, x **2)
ax2.plot(x, x /3)
ax3.plot(x, x **4)
ax4.plot(x, x *5)
```

```python
# Subplot 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                            ncols=2,
                                            figsize = (10,5))

ax1.plot(x, x /2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(coldrink_prices.keys(), coldrink_prices.values())
```
