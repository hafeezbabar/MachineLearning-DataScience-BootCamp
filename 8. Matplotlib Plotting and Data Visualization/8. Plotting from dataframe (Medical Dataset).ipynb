 - When plotting something more advanced , use the matplotlib method

```python
## Method from scratch
fig, ax  = plt.subplots(figsize = (10,6))


# Plot the data
scatter = ax.scatter(x = age_over_50["age"],
                    y = age_over_50["chol"],
                    c = age_over_50["target"])

# Customize the Plot
ax.set(title="Heart Disease Visualization",
      xlabel = "Age",
      ylabel = "Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target")
```

```python
## Add a horizontal line in Graph

## Method from scratch
fig, ax  = plt.subplots(figsize = (10,6))


# Plot the data
scatter = ax.scatter(x = age_over_50["age"],
                    y = age_over_50["chol"],
                    c = age_over_50["target"])

# Customize the Plot
ax.set(title="Heart Disease Visualization",
      xlabel = "Age",
      ylabel = "Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target")

## Add a horizontal line in Graph
ax.axhline(age_over_50["chol"].mean(), 
          linestyle = '--')
```

```python
### subplots of few columns
fig, (ax0, ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize = (10,10))
# Add Data to ax0
scatter = ax0.scatter(x = age_over_50["age"],
                     y = age_over_50["chol"],
                     c = age_over_50["target"])

# Customize ax0
ax0.set(title= " Heart Disease Visualization",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y = age_over_50["chol"].mean(),
           linestyle = "--")

## Add data to ax1
scatter = ax1.scatter(x = age_over_50["age"],
                     y = age_over_50["thalach"],
                     c = age_over_50["target"])

# Customize  ax1
ax1.set(title= " Heart Disease Visualization",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y = age_over_50["thalach"].mean(),
           linestyle = "--")


# ADd a title to Figure, 
fig.suptitle("Heart Disease Analysis", fontsize = 20)
```
